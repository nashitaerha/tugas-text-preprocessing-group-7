# **Preprocessing**

```python
import pandas as pd

df = pd.read_csv('review_cam_scanner.csv')

df.to_csv('review_2.csv', index=False)

print("File saved as 'reviews_2.csv'.")
```

Removal of Emojis

```python
!pip install emoji
import re
import emoji

df = pd.read_csv('review_2.csv')

def emoji_removal(content):
  content = emoji.replace_emoji(content, replace='')
  return content

df['content'] = df['content'].apply(lambda x: emoji_removal(x))
df.head(50)
```

```python
df.to_csv("cek_emoji.csv", index = False)
```

Lower Casing

```python
df['content'] = df['content'].str.lower()
df.head(50)
```

Removal of Emoticon

```python
import pandas as pd
import re

EMOTICONS = {
    u":-\)":"Happy face or sm\ley",
    u":\)":"Happy face or smitéy",
    u":-D":"Laughing, big grin or laugh with glasses",
    u":D":"Laughing, big grin or laugh with glasses",
    u"8-D":"Laughing, big grin or laugh with glasses",
    u":-\)\)":"Very happy",
    u":-\(":"Frown, sad, andry or pouting"}

def remove_emoticons(text):
    if not isinstance(text, (str, bytes)):
        text = str(text)
    emoticon_pattern = re.compile(u'(' + u'|'.join(k for k in EMOTICONS) + u')')
    return emoticon_pattern.sub(r'', text)

df['content'] = df['content'].apply(lambda text: remove_emoticons(text))
df.head(50)
```

Removal of Punctuations

```python
def remove_punctuation(text):
    import string
    text = str(text)
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
```

```python
df['content'] = df['content'].apply(lambda text: remove_punctuation(text))
df.head(50)
```

Removal of Stopwords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def remove_stopwords(text):
    stop_words = set(stopwords.words('indonesian'))
    return ' '.join([word for word in text.split() if word not in stop_words])
```

```python
df['content'] = df['content'].apply(lambda text: remove_stopwords(text))
df.head(50)
```

Removal of Frequent Words

```python
from collections import Counter
cnt = Counter()
for text in df.content.values:
    if isinstance(text, str):
        for word in text.split():
            cnt[word] += 1
cnt.most_common(10)
```

```python
from collections import Counter
cnt= Counter()
for text in df.content.values:
    if isinstance(text, str):
        for word in text.split():
            cnt[word] += 1

cnt.most_common(10)
FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

df['content'] = df['content'].apply(lambda text: remove_freqwords(text))
df.head(50)
```

```python
from IPython.display import display

df = df[df['content'] != '']
display(df)
```

# **Pelabelan**

```python
def pelabelan(score):
  if score < 3:
    return 'Negatif'
  elif score == 4:
    return 'Positif'
  elif score == 5:
    return 'Positif'

df['Label'] = df['score'].apply(pelabelan)
df.head(50)
```

# **Tokenizing**

```python
from nltk.tokenize import word_tokenize
```

```python
from nltk.tokenize import RegexpTokenizer

regexp = RegexpTokenizer('\w+')

# Convert the 'content' column to strings before applying the tokenizer
df['content_token'] = df['content'].astype(str).apply(regexp.tokenize)

df.head(50)
```

```python
import pandas as pd
import numpy as np
```

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.intsig.camscanner',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=1000,
    filter_score_with=None
)
```

```python
data = pd.DataFrame(np.array(result), columns=['review'])
data = data.join(pd.DataFrame(data.pop('review').tolist()))
data
```

```python
scrappeddata1 = data[['userName','content','score','at']]
sorteddata = scrappeddata1.sort_values(by='at', ascending=False)
sorteddata.head()
```

```python
sorteddata['Year'] = sorteddata['at'].dt.year
sorteddata['Month'] = sorteddata['at'].dt.month
sorteddata['Day'] = sorteddata['at'].dt.day
sorteddata
```

```python
df = sorteddata[['userName','content','score','Year','Month','Day']]
df
```

```python
df.to_csv("review_cam_scanner.csv", index = False)
```
